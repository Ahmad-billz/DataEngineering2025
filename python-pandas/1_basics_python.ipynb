# Basics of Maths Operations

```python
first_num = 5
second_num = 6
print(first_num + second_num)
print(first_num - second_num)
print(first_num / second_num)
print(first_num * second_num)
print(first_num ** second_num)  # (a, b) a raised to the power of b (3, 4) 3 to the power of 4
print(first_num % second_num) # modulus - remainder
print(first_num // second_num) #floor division operation nearest down whole number
```

## Error Handling

```python
x = 10
y = 0
try: 
    print(x/y)
except ZeroDivisionError as err:
    print(f"The user tried a number which is zero for division.")
except Exception as e:
    print(f"Error found : {e}")
```

```python
x = 10
y = 0
try: 
    print(x/y)

except Exception as e:
    print(f"Error found : {e}")
except ZeroDivisionError as err:
    print("The user tried a number which is zero for division.")
finally:
    print("Lets go for the other task.")
```

```python
def check_number(age):
    try:
        if age < 0:
            raise ValueError("Number is incorrect")
        return f"{age} is valid"
    except Exception as e:
        print(f"Error : {e}")
```

```python
age = int (input("Enter your age no: "))
check_number(age)
```

## Generator, iterator, yield

```python
iterable = [1,2,3,4,5,]
iterator = iter(iterable)
print(next(iterator))
```

```python
list(range (1,6))
```

```python
def numbers():
    for i in range (1, 6):
        yield i

gen = numbers()

print(next(gen))
print(next(gen))
print(next(gen))

print("_________")

for n in gen:
    print(n)
```
